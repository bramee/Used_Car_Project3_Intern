```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests
```

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

url = "https://www.theaa.com/used-cars/supermini-for-sale"
driver.get(url)
time.sleep(6)
```

```python
brand = []

for i in range(1,500):
    name = driver.find_elements_by_xpath("//div[@class= 'vl-list']/div/h3")
    for i in name:
        brand.append(i.text)
driver.find_element_by_xpath("//div[@class= 'container']/div/button[2]").click()
time.sleep(2)

brand
```

```python
price = []

for i in range(1,500):
    pri = driver.find_elements_by_xpath("//div[@class= 'vl-list']/div/div/div/div/strong")
    for i in pri:
        price.append(i.text)
driver.find_element_by_xpath("//div[@class= 'container']/div/button[2]").click()
time.sleep(2)

price
```

```python
year = []

for i in range(1,500):
    yea = driver.find_elements_by_xpath("//div[@class= 'vl-list']/div/div/div/ul/li[1]")
    for i in yea:
        year.append(i.text)
driver.find_element_by_xpath("//div[@class= 'container']/div/button[2]").click()
time.sleep(2)

year
```

```python
fuel = []

for i in range(1,500):
    fue = driver.find_elements_by_xpath("//div[@class= 'vl-list']/div/div/div/ul/li[5]")
    for i in fue:
        fuel.append(i.text)
driver.find_element_by_xpath("//div[@class= 'container']/div/button[2]").click()
time.sleep(2)

fuel
```

```python
trans = []

for i in range(1,500):
    tra = driver.find_elements_by_xpath("//div[@class= 'vl-list']/div/div/div/ul/li[7]")
    for i in tra:
        trans.append(i.text)
driver.find_element_by_xpath("//div[@class= 'container']/div/button[2]").click()
time.sleep(2)

trans
```

```python
run = []

for i in range(1,500):
    runn = driver.find_elements_by_xpath("//div[@class= 'vl-list']/div/div/div/ul/li[3]")
    for i in runn:
        run.append(i.text)
driver.find_element_by_xpath("//div[@class= 'container']/div/button[2]").click()
time.sleep(2)

run
```

```python
print(len(brand),len(price),len(year),len(fuel),len(trans),len(run))
```

```python
Usedcar = pd.DataFrame({'Brand': brand,'Price':price,'Year':year,'Fuel':fuel,'Transmission':trans,'Running KM':run})
Usedcar.to_excel("UsedCar.xlsx")
```
